```python
import pandas as pd
import matplotlib.pyplot as plt
hpgres = pd.read_csv('restaurant/hpg_reserve.csv')
hpgres.tail()
```

```python
import csv
import numpy as np
hpg_id = hpgres.hpg_store_id;
hpg_date = hpgres.visit_datetime;
list_id = [];
list_date = []; # 无重复的日期
list_res_date = []; # 所有的日期
# get date
for i in range(0,len(hpg_date)):
    split_date = hpg_date[i].split(' ');
    split_date2 = split_date[0].split('-');
    list_res_date.append("date_"+split_date2[0]+"_"+split_date2[1]+"_"+split_date2[2]);
    if("date_"+split_date2[0]+"_"+split_date2[1]+"_"+split_date2[2] not in list_date):
        list_date.append("date_"+split_date2[0]+"_"+split_date2[1]+"_"+split_date2[2]);
'''
for i in range(0,len(hpg_id)):
    if(hpg_id[i] not in list_id):
        list_id.append(hpg_id[i]);        
'''
```

```python
import time
start = time.clock();
day_vis = np.zeros((len(list_id),len(list_date)))
for k in range(0, len(hpg_id)):
    timeConsumed = (time.clock() - start);
    print("正在计算第",k,"个餐厅的日访问量, 已耗时", timeConsumed, "秒");
    i = 0;
    j = 0;
    while (hpg_id[k] != list_id[i]) :
        i = i + 1;
        # throw exception
        if (i > len(list_id)):
            print("error i");
    while (list_res_date[k] != list_date[j]):
        j = j + 1;
        # throw exception
        if (j > len(list_date)):
            print("error i");
    day_vis[i,j] = hpgres.reserve_visitors[k] + day_vis[i,j];
np.savetxt('hpg_daily_res_visitors.csv', day_vis, delimiter = ',') 
```
