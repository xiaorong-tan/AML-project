```python
import pandas as pd
import matplotlib.pyplot as plt
airres = pd.read_csv('restaurant/air_reserve.csv')
airres.head(20)
```

```python
import csv
import numpy as np
air_id = airres.air_store_id;
air_date = airres.visit_datetime;
list_id = [];
list_date = []; # 无重复的日期
list_res_date = []; # 所有的日期
# get date
for i in range(0,len(air_date)):
    string_date = air_date[i];
    split_date = string_date.split(' ');
    list_res_date.append(split_date[0]);
    if(split_date[0] not in list_date):
        list_date.append(split_date[0]);

for i in range(0,len(air_id)):
    if(air_id[i] not in list_id):
        list_id.append(air_id[i]);
        
```

```python
np.savetxt('air_daily_res_visitors.csv', day_vis, delimiter = ',') 
```
